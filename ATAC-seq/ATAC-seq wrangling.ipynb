# Filtering out low variance peaks

Peaks with low variance represent chromatin landscape regions with low variation across cell types. These can be left out of our further analysis as they give us little input to differing regulatory landscapes between cell lines.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Converting signal columns to float32 for memory efficiency
filtered_all_ATAC = filtered_all_ATAC.copy()
filtered_all_ATAC.loc[:, cell_types_to_keep] = filtered_all_ATAC[cell_types_to_keep].astype('float32')
filtered_all_ATAC.loc[:, 'variance'] = filtered_all_ATAC[cell_types_to_keep].var(axis=1)


# 2. Computing per-peak variance across the selected cell types
sample_cols = cell_types_to_keep
filtered_all_ATAC['variance'] = filtered_all_ATAC[sample_cols].var(axis=1)

# 3. Inspecting the distribution of variances to choose a cutoff

plt.hist(filtered_all_ATAC['variance'], bins=100)
plt.xlabel('Peak variance across cell types')
plt.ylabel('Frequency')
plt.show()
```

Using log-scaling as the previous graph is pretty stacked up.

```python
# Add a small constant so zeros don’t turn into –∞
x = np.log10(filtered_all_ATAC['variance'] + 1e-3)

plt.figure()
plt.hist(x, bins=100)
plt.xlabel('log10(variance + 1e-3)')
plt.ylabel('Frequency')
plt.title('Distribution of log-transformed peak variances')
plt.show()
```

Method chosen for finding a cutoff for filtering out low variance peaks. This was the most meaningful one.

- MAD flags peaks whose log-variance is more than 2× the median absolute deviation above the median.

```python
import numpy as np
import pandas as pd

filtered_all_ATAC = pd.read_csv("filtered_ATAC_abT_Tact_Stem.csv")

# 2. Make sure 'variance' exists:
if 'variance' not in filtered_all_ATAC:
    meta_cols   = filtered_all_ATAC.columns[:8].tolist()
    sample_cols = [c for c in filtered_all_ATAC.columns if c not in meta_cols]
    # cast to float32 to save memory
    filtered_all_ATAC[sample_cols] = filtered_all_ATAC[sample_cols].astype('float32')
    filtered_all_ATAC['variance'] = filtered_all_ATAC[sample_cols].var(axis=1)

# 3. Compute the MAD-based cutoff (k=2)
logv      = np.log10(filtered_all_ATAC['variance'] + 1e-6)
med       = np.median(logv)
mad       = np.median(np.abs(logv - med))
k         = 2
cut_log   = med + k * mad
cut_var   = 10 ** cut_log

print(f"MAD variance cutoff: {cut_var:.3f} (log10 = {cut_log:.3f})")

# 4. Subset
high_var = filtered_all_ATAC[filtered_all_ATAC['variance'] > cut_var].copy()

print(f"Retained {len(high_var)} / {len(filtered_all_ATAC)} peaks "
      f"({len(high_var)/len(filtered_all_ATAC)*100:.1f}%)")

```

# Hexbin Plot

Using a hexbin plot can be helpful to analyze how variance and another metric correlate. By doing so we can have an even more meaningful and sensible cutoff for the ATAC peaks.

```python
import matplotlib.pyplot as plt

# Assumes filtered_all_ATAC is in memory and has a 'variance' column

plt.figure(figsize=(6,5))
hb = plt.hexbin(
    filtered_all_ATAC['_-log10_bestPvalue'],
    filtered_all_ATAC['variance'],
    gridsize=60,
    yscale='log',
    mincnt=1
)
plt.colorbar(hb, label='Count of peaks')
plt.xlabel('-log10(best P-value)')
plt.ylabel('Variance (per peak)')
plt.title('Hexbin: Variance vs. –log10(best P-value)')
plt.show()
```

This basically shows us that there are high variance peaks with low significance/confidence scores. Might also be useful to filter these out.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Recompute MAD-based cutoffs ---
# log-variance
log_v = np.log10(filtered_all_ATAC['variance'] + 1e-6)
med_v = np.median(log_v)
mad_v = np.median(np.abs(log_v - med_v))
k     = 2
cut_v_log   = med_v + k * mad_v
cut_var_mad = 10 ** cut_v_log

# -log10 P-value
lp    = filtered_all_ATAC['_-log10_bestPvalue'].values
med_lp = np.median(lp)
mad_lp = np.median(np.abs(lp - med_lp))
cut_lp = med_lp + k * mad_lp

print(f"Variance cutoff = {cut_var_mad:.3f} (log10 = {cut_v_log:.3f})")
print(f"-log10P cutoff  = {cut_lp:.3f}")

# --- 2. Plot hexbin with threshold lines ---
plt.figure(figsize=(6,5))
hb = plt.hexbin(
    filtered_all_ATAC['_-log10_bestPvalue'],
    filtered_all_ATAC['variance'],
    gridsize=60,
    yscale='log',
    mincnt=1,
    cmap='viridis'
)
plt.colorbar(hb, label='Peak count')

# vertical line for -log10P threshold
plt.axvline(cut_lp, color='red', linestyle='--', linewidth=1, label=f"-log10P > {cut_lp:.1f}")
# horizontal line for variance threshold
plt.axhline(cut_var_mad, color='orange', linestyle='--', linewidth=1, label=f"variance > {cut_var_mad:.1f}")

plt.xlabel('-log10(best P-value)')
plt.ylabel('Variance')
plt.title('Variance vs. –log10P with MAD thresholds')
plt.legend(frameon=False)
plt.tight_layout()
plt.show()

# --- 3. Subset peaks passing both criteria ---
high_var_sig = filtered_all_ATAC[
    (filtered_all_ATAC['variance']   > cut_var_mad) &
    (filtered_all_ATAC['_-log10_bestPvalue'] > cut_lp)
].copy()

# --- 4. Save to CSV ---
high_var_sig.to_csv("ATAC_high_var.csv", index=False)
print(f"Saved {len(high_var_sig)} peaks to ATAC_high_var.csv ({len(high_var_sig)/len(filtered_all_ATAC)*100:.2f}% retained)")
```

With 14.8% of peaks retained, analyzing the data is easier, faster and cleaner. Even though we have less data, it is more meaningful.
