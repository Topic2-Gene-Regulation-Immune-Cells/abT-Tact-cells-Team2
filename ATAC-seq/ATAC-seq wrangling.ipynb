```python
import pandas as pd
df = pd.read_csv("ATAC-seq data.csv")
print("All columns")
print(df.columns)
```

```python
meta_cols = df.columns[:8].tolist()

# select only relevant cells
cell_types_to_keep = [
    "preT.DN1.Th", "preT.DN2a.Th", "preT.DN2b.Th", "preT.DN3.Th",
    "T.DN4.Th", "T.ISP.Th", "T.DP.Th", "T.4.Th", "T.8.Th",
    "T.4.Nve.Sp", "T.4.Nve.Fem.Sp", "T.4.Sp.aCD3+CD40.18hr", "T.8.Nve.Sp",
    "Treg.4.25hi.Sp", "Treg.4.FP3+.Nrplo.Co",
    "T8.TN.P14.Sp", "T8.TE.LCMV.d7.Sp", "T8.MP.LCMV.d7.Sp", 
    "T8.IEL.LCMV.d7.SI",   
    "T8.Tcm.LCMV.d180.Sp", "T8.Tem.LCMV.d180.Sp",
    "NKT.Sp", "NKT.Sp.LPS.3hr", "NKT.Sp.LPS.18hr", "NKT.Sp.LPS.3d",
    "LTHSC.34-.BM", "LTHSC.34+.BM", "STHSC.150-.BM", "MPP4.135+.BM"
]

# now subset including metadata to include peak IDs/chr/etc.
filtered_all_ATAC = df[ meta_cols + cell_types_to_keep ]

# save & check
filtered_all_ATAC.to_csv("filtered_ATAC_abT_Tact_Stem.csv", index=False)
print("Filtered shape:", filtered_all_ATAC.shape)
print("Columns kept:", filtered_all_ATAC.columns.tolist())
```

```python
# show as table

filtered_all_ATAC.head()
```

```python
# melt subset so that each row is one peak-sample pair with SampleID and its corresponding singnal
# it's easier for later analysis

tidy = filtered_all_ATAC.melt(
    id_vars=meta_cols,
    value_vars=filtered_all_ATAC,
    var_name="Sample ID",
    value_name="Signal",
)
tidy.head()
```

```python
# export for further analysis

tidy.to_csv( "refined_ATAC.csv", index=False)
print( "Tidy ATAC shape:", tidy.shape)
print(tidy.head())
```
